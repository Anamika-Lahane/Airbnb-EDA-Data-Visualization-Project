**Task 5: Data Analysis**

**Univariate Analysis**

```python
#OUTLIERS
#df=data[data['price']<100000]
df=data[data['price']<1500]
sns.boxplot(data=df, x='price')
```

```python
#Univariate Analysis
plt.figure(figsize=(8, 5))
sns.histplot(df['price'], bins=60, kde=True)
plt.title('PRICE DISTRIBUTION')
plt.show()
```

**CATEGORICAL ANALYSIS**

```python
#What type of rooms are more common?
sns.countplot(data=df ,x='room_type', hue='room_type')
plt.title("Room Type Distribution")
plt.show()
```

```python
#Which neighborhoods are most listed?
df['neighbourhood_group'].value_counts().plot(kind='bar')
plt.title('Most Listed Neighbourhood groups')
plt.show()
```

```python
data.columns
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='availability_365')
plt.title('availability_365 Distribuition')
plt.show()
```

```python
df = df.dropna(subset=['beds'])

df.isnull().sum()
```

```python
#CREATING NEW COLUMN PRICE PER BED
data['price_per_bed']=data['price']/data['beds']
data.head()
```

**Bi Variable Analysis One variable depenency in another variable**

```python
#Price by Room Type
sns.barplot(data=df, x='room_type', y='price', hue='room_type')
plt.title('Price by Room Type')
plt.show()
```

```python
#Availability vs Rating
df['rating'] = pd.to_numeric(df['rating'], errors='coerce') #to remove errors on y axis

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='availability_365', y='rating', hue='availability_365')
plt.title('Availability vs Rating')
plt.show()
```

```python
#Reviews vs Price
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='number_of_reviews')
plt.title('Reviews vs Price')
plt.show()
```

```python
# price dependency on neighbourhood
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
plt.title('Price dependency on neighbourhood')
plt.show()
```

```python
#Geospatial Analysis
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='neighbourhood_group')
plt.title('Listing by Location')
plt.show()
```

```python

# Extract year from last_review
df['review_year'] = df['last_review'].dt.year

# Plot count of reviews by year, colored by room_type
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='review_year', hue='room_type')
plt.title("Review Count per Year by Room Type")
plt.xticks(rotation=60)
plt.show()

```

```python
# heat map - correlation of one variable with others for numerical column

corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr

plt.figure(figsize=(8, 5))
sns.heatmap(data=corr, annot=True)
plt.title('correlation')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.pairplot(data=df, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue='room_type')
plt.show()
```
